```python
import torch as tn
import torchtt as tntt
import TTCME
import matplotlib.pyplot as plt 
import datetime
import numpy as np
import pickle 

tn.set_default_tensor_type(tn.DoubleTensor)
```

```python
r1m = TTCME.ChemicalReaction(['S'],'->S', 5.0)
r2m = TTCME.ChemicalReaction(['S'],'S->', 0.2)


mdl = TTCME.ReactionSystem(['S'],[r1m, r2m])

IC = [0]
N = [64]
```

```python
Tend = 100



reaction_time,reaction_jumps,reaction_indices = mdl.ssa_single(IC, Tend)
```

```python
plt.figure()
plt.plot(np.repeat(reaction_time,2)[1:],np.repeat(reaction_jumps[:,0],2)[:-1])
plt.legend(['S'])
plt.xlabel(r'$t$')
plt.ylabel(r'#individuals')
import tikzplotlib
tikzplotlib.save('./birth_death_sample.tex')
```

```python
N = 100
plt.figure()
for _ in range(N):
    reaction_time,reaction_jumps,reaction_indices = mdl.ssa_single(IC, Tend)
    plt.plot(np.repeat(reaction_time,2)[1:],np.repeat(reaction_jumps[:,0],2)[:-1],'b',linewidth=0.1)
# plt.legend(['S'])
plt.xlabel(r'$t$')
plt.ylabel(r'#individuals')
import tikzplotlib
tikzplotlib.save('./birth_death_sample.tex')
```
